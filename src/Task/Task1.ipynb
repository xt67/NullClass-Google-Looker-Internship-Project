```python
connection = pymysql.connect(
    host='localhost',
    user='root',
    password='July@2019',
    db='sales'
)
query = """
SELECT customer_id, registered_date, is_gross, is_valid, is_net, order_date
FROM order_detail o
JOIN customer_detail c ON o.customer_id = c.id
"""
df = pd.read_sql(query, connection)
```

```python
df['order_date'] = pd.to_datetime(df['order_date']) 
cs_gross_data = df.loc[
    (df['is_gross'] == 1) &
    (df['is_valid'] == 0) &
    (df['is_net'] == 0) &
    (df['order_date'].dt.year == 2022),
    ['customer_id', 'registered_date']
].drop_duplicates()
cs_gross_data
```

```python
cs_gross_data = cs_gross_data.reset_index(drop=True)
cs_gross_data
```
